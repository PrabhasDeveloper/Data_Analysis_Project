# dataset:-govtdataset.csv



```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import math

df=pd.read_csv("govtdataset.csv")
df
```

# Loaded the dataset: Read the CSV file (`govtdataset.csv`) using `pd.read_csv()` and displayed the dataset.


```python
df.head(3)
```

```python
df.info()
```

# Checked dataset information: Displayed column types and non-null counts using `df.info()`.


```python
df["DDate"] = pd.to_datetime(df["DDate"], errors='coerce')
df
```

```python
df = df.drop_duplicates()
df
```

# Removed duplicate entries and Cleaned the dataset by dropping duplicate rows.


```python
df.fillna(0, inplace=True)
df
```

```python
for col in ["Arrivals", "Minimum", "Maximum", "Model", "Valuation", "MarketFee"]:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), np.nan, df[col])

df
```

# **Applied Outlier Removal (IQR Method):** Removed outliers from columns - Arrivals, Minimum, Maximum, Model, Valuation, and MarketFee.


df.dropna(inplace=True)
df

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('govtdataset.csv')
subset = data[['CommName', 'Minimum', 'Maximum']].dropna().head(10)

plt.figure(figsize=(10, 6))
x = range(len(subset))
plt.bar(x, subset['Minimum'], color='beige', label='Minimum', alpha=0.8)
plt.bar(x, subset['Maximum'], color='tan', label='Maximum', alpha=0.8)
plt.xticks(x, subset['CommName'], rotation=45, ha='right')
plt.title('Minimum and Maximum Prices of Commodities', fontsize=14)
plt.ylabel('Price')
plt.xlabel('Commodity')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(subset['Maximum'], labels=subset['CommName'], autopct='%1.1f%%',
        colors=['#f5deb3', '#f0e68c', '#d2b48c', '#deb887', '#cdb79e',
                '#ffe4c4', '#e0c2a2', '#ecd9c6', '#d8bfd8', '#c9b59a'])
plt.title('Proportion of Maximum Prices', fontsize=14)
plt.tight_layout()
plt.show()
```
